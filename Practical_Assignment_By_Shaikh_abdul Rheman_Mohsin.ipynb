```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\admin\Ai Boot Camp\ipl_matches_smalldata.csv")
print("Columns in dataset:", df.columns)
player_runs = df.groupby("batter")["batsman_runs"].sum().reset_index()
top_10_players = player_runs.sort_values(by="batsman_runs", ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_players["batter"], top_10_players["batsman_runs"], color='blue')
plt.xlabel("Players", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.title("Top 10 Players by Runs in IPL", fontsize=14)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
text = """WrestleMania is one of the biggest annual events in professional wrestling, organized by WWE. It features the top WWE Superstars competing in high-stakes matches, including championship bouts, grudge matches, and special attractions. The event is known for its grand entrances, celebrity appearances, and historic moments. WrestleMania takes place in large stadiums and is watched by millions of fans worldwide. The goal for wrestlers is to defeat their opponents through pinfalls, submissions, count-outs, or disqualifications. The event often marks the culmination of long-running rivalries and showcases the best of sports entertainment.
"""
print(text)
words = word_tokenize(text.lower()) 
print(words)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
word_freq = Counter(filtered_words)
print(word_freq)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') 
plt.show()
```
