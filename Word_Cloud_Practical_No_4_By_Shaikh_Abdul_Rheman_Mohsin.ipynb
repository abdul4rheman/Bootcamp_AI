```python

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = """
WrestleMania is one of the biggest annual events in professional wrestling, organized by WWE. It features the top WWE Superstars competing in high-stakes matches, including championship bouts, grudge matches, and special attractions. The event is known for its grand entrances, celebrity appearances, and historic moments. WrestleMania takes place in large stadiums and is watched by millions of fans worldwide. The goal for wrestlers is to defeat their opponents through pinfalls, submissions, count-outs, or disqualifications. The event often marks the culmination of long-running rivalries and showcases the best of sports entertainment.
"""
print(text)
```

```python
words = word_tokenize(text.lower())  # Convert to lowercase for uniformity
print(words)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
```

```python
word_freq = Counter(filtered_words)
print(word_freq)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') 
plt.show()
```
